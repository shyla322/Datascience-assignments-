# outliers checking

```python

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,8))
sns.boxenplot(df1)
```

# Data Visualiztion

```python
# Boxplot
for i in df1:
    plt.boxplot(df1[i])
    plt.title(i)
    plt.show()
```

```python
#Histogram
```

```python
for i in df1:
    sns.histplot(df1[i], kde = True,
                line_kws = {'linestyle':'dashed',
                         'linewidth':'2'}).lines[0].set_color('red')
    plt.title(i)
    plt.show()
```

```python
A = []
B = []

for i in df1.columns:
    mean_1 = np.mean(df1[i])
    if mean_1 > 5:
        A.append(i)
        B.append(mean_1)
plt.pie(B,labels=A,autopct='%1.0f%%')
```

```python
# Heat Map
sns.heatmap(data=df1)
```
